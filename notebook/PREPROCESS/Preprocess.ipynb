```python
import unicodedata
# Normalize text
def normalize_unicode(text):
    return unicodedata.normalize('NFC', text) if isinstance(text, str) else text
```

```python
def process_excel_file(file_path):
    # Read the Excel file
    df = pd.read_excel(file_path)

    # Apply transformations: lowercase, strip whitespace, normalize Unicode
    df = df.applymap(lambda x: normalize_unicode(x.strip().lower()) if isinstance(x, str) else x)

    # Remove spaces from column names
    df.columns = [col.replace(' ', '') for col in df.columns]

    return df
```

```python
def check_duplicates(df, subset):
    # Find duplicates based on the subset
    duplicates = df[df.duplicated(subset=subset, keep=False)]

    if not duplicates.empty:
        print("Duplicates found:")
        return duplicates
    else:
        print("No duplicates found.")
        return None
```

```python
# Check Null precentage in data

def Null_percentage(df):
    null_percentage = (df.isnull().mean() * 100).round(2)
    null_percentage_df = pd.DataFrame({'Column': null_percentage.index, 'NullPercentage': null_percentage.values})
    return null_percentage_df
```

```python
# Change value in dataframe on target column

def Change_value(df_in, value, target, col):
    df_in[col] = df_in[col].replace(value, target)

    mask = df_in[col].str.contains('|'.join(value), case=False)
    df_in.loc[mask, col] = target
```
